```python
#Más métodos de plots
import matplotlib.pyplot as plt
import seaborn as sns

def plot_bars_action(dic_results):####de aqui para abaixo hai que miralo todo
    rounds = len(dic_results['azul']) ## Aquí escribín azul mais pode ser calquera
    final_azul = [0]*4
    flag = [0]*(len(dic_results['azul']))
    for k in range(4):
        l1 = [dic_results['azul'][i][k] for i in range(len(dic_results['azul']))]
        flag = [a+b for a, b in zip(flag, l1)]
        final_azul[k] = flag
    df_azul =pd.DataFrame({'round':range(rounds),'=,0':final_azul[0],'=,1':final_azul[1],'!=,0':final_azul[2],'!=,1':final_azul[3]})

    final_amarillo = [0]*4
    flag = [0]*(len(dic_results['amarillo']))
    for k in range(4):
        l1 = [dic_results['amarillo'][i][k] for i in range(len(dic_results['amarillo']))]
        flag = [a+b for a, b in zip(flag, l1)]
        final_amarillo[k] = flag    
    df_amarillo =pd.DataFrame({'round':range(rounds),'=,0':final_amarillo[0],'=,1':final_amarillo[1],'!=,0':final_amarillo[2],'!=,1':final_amarillo[3]})        

    df_total = df_azul + df_amarillo
    df_total['round'] = df_total['round'].apply(lambda x:x//2)
    df_return = [df_azul,df_amarillo,df_total]
    names = ['azul','amarillo','total']
    for i,df in enumerate(df_return):
        sns.set_color_codes("pastel")
        f, ax = plt.subplots(figsize=(12, 8))
        sns.barplot(x='round',y='!=,1',data = df,color = 'b',label = '!=,1').set_title(names[i]+',group '+str(group))
        sns.barplot(x='round',y='!=,0',data = df,color = 'r',label = '!=,0')
        sns.barplot(x='round',y='=,1',data = df,color = 'c',label = '=,1')
        sns.barplot(x='round',y='=,0',data = df,color = 'g',label = '=,0')
        ax.legend(ncol=4, loc="best", frameon=True)
        ax.set_xlabel('round')
        ax.set_ylabel('options')
        plt.savefig(os.path.join(start_point,join_name)+'/bar_action_'+str(names[i])+'_group_'+str(group)+'_'+str(study_sessions[ij])+'_'+str(join_name)+'.eps')
        plt.close(f)
#ax.title(names[i])

        
    return 

def plot_bars_success(dic_results):####de aqui para abaixo hai que miralo todo
    rounds = len(dic_results['azul']) ## Aquí escribín azul mais pode ser calquera
    final_azul = [0]*4
    flag = [0]*(len(dic_results['azul']))
    for k in range(4):
        l1 = [dic_results['azul'][i][k] for i in range(len(dic_results['azul']))]
        flag = [a+b for a, b in zip(flag, l1)]
        final_azul[k] = flag
    df_azul =pd.DataFrame({'round':range(rounds),'=,coord':final_azul[0],'=,uncoord':final_azul[1],'!=,coord':final_azul[2],'!=,uncoord':final_azul[3]})

    final_amarillo = [0]*4
    flag = [0]*(len(dic_results['amarillo']))
    for k in range(4):
        l1 = [dic_results['amarillo'][i][k] for i in range(len(dic_results['amarillo']))]
        flag = [a+b for a, b in zip(flag, l1)]
        final_amarillo[k] = flag    
    df_amarillo =pd.DataFrame({'round':range(rounds),'=,coord':final_amarillo[0],'=,uncoord':final_amarillo[1],'!=,coord':final_amarillo[2],'!=,uncoord':final_amarillo[3]})        

    df_total = df_azul + df_amarillo
    df_total['round'] = df_total['round'].apply(lambda x:x//2)
    df_return = [df_azul,df_amarillo,df_total]
    names = ['azul','amarillo','total']
    for i,df in enumerate(df_return):
        fig, ax = plt.subplots(figsize=(12, 8))
        sns.set_color_codes("pastel")
        sns.barplot(x='round',y='!=,uncoord',data = df,color = 'salmon',label = '!=,uncoord').set_title(names[i]+',group '+str(group))
        sns.barplot(x='round',y='!=,coord',data = df,color = 'red',label = '!=,coord')
        sns.barplot(x='round',y='=,uncoord',data = df,color = 'palegreen',label = '=,uncoord')
        sns.barplot(x='round',y='=,coord',data = df,color = 'green',label = '=,coord')
        ax.legend(ncol=4, loc="best", frameon=True)
        ax.set_xlabel('round')
        ax.set_ylabel('options')
        plt.savefig(os.path.join(start_point,join_name)+'/bar_success_'+str(names[i])+'_group_'+str(group)+'_'+str(study_sessions[ij])+'_'+str(join_name)+'.eps')
        plt.close(fig)
    return 

def bias_dic(dataframes_saved):
    dic_bias = {'amarillo':532,'azul':594}
    total_len = sum([len(dataframes_saved[i]) for i in range(len(dataframes_saved))])
    distancias_igual = [0 for ij in range(total_len)]
    distancias_diff =  [0 for ij in range(total_len)]
    k = 0
    for players_info in dataframes_saved:
        for i in range(len(players_info)):
            for mode in ['igual','diferente']:
                if mode == 'igual':
                    distancias_igual[k] = abs(dic_bias[players_info[i][0]['Marker']] - players_info[i][0]['Bias'])
                else: 
                    if players_info[i][0]['Marker'] == 'amarillo': maskarita = 'azul'
                    if players_info[i][0]['Marker'] == 'azul': maskarita = 'amarillo'
                    distancias_diff[k] = abs(dic_bias[maskarita] - players_info[i][0]['Bias'])
            k+=1


    nearer = [(distancias_diff[i]-distancias_igual[i])/(distancias_igual[i]+distancias_diff[i]) for i in range(len(distancias_igual))]
    bias_plots = pd.DataFrame({'distance to same marker reference':distancias_igual,
                              'distance to different marker reference':distancias_diff,
                              'trust': nearer
                              })

    fig = plt.figure(figsize=(12,8))
    sns.displot(bias_plots['trust'],bins = 10)
    plt.savefig(os.path.join(start_point,join_name)+'/trust_'+str(join_name)+'.eps')
    plt.close(fig)


    g = sns.JointGrid(data=bias_plots, x="distance to same marker reference",
                      y="distance to different marker reference")
    fig_2 = plt.figure(figsize=(12,8))
    g.plot(sns.scatterplot, sns.histplot)
    g.savefig(os.path.join(start_point,join_name)+'/bias_distribution_'+str(join_name)+'.eps')
    plt.close(fig_2)

    return 
```

```python
def bar_action(dataframes_saved):
    for ij,players_info in enumerate(dataframes_saved):  
        for group in range(1,int(dataframes_saved[ij][-1][0]['id_in_subsession']+1)):
            players_info_group = [players_info[i] for i in range(len(players_info)) if players_info[i][0]['id_in_subsession'] == group]
            results_azul = [[0 for ij in range(4)] for j in range(len(dataframes_saved[0][1][1]))]
            results_amarillo = [[0 for ij in range(4)] for j in range(len(dataframes_saved[0][1][1]))]
            for player in players_info_group:
                for i in range(len(player[1])):
                    if player[0]['Marker'] == player[1]['marker_partner'].iloc[i]:
                        k = 0
                    else:
                        k = 2      
                    if player[0]['Marker'] == 'azul':
                        results_azul[i][k+int(player[1]['acción'].iloc[i])-1] += 1
                    elif player[0]['Marker'] == 'amarillo':
                        results_amarillo[i][k+int(player[1]['acción'].iloc[i])-1] += 1
                dic_results = {'azul':results_azul,'amarillo':results_amarillo}


            rounds = len(dic_results['azul']) ## Aquí escribín azul mais pode ser calquera
            final_azul = [0]*4
            flag = [0]*(len(dic_results['azul']))
            for k in range(4):
                l1 = [dic_results['azul'][i][k] for i in range(len(dic_results['azul']))]
                flag = [a+b for a, b in zip(flag, l1)]
                final_azul[k] = flag
            df_azul =pd.DataFrame({'round':range(rounds),'=,0':final_azul[0],'=,1':final_azul[1],'!=,0':final_azul[2],'!=,1':final_azul[3]})

            final_amarillo = [0]*4
            flag = [0]*(len(dic_results['amarillo']))
            for k in range(4):
                l1 = [dic_results['amarillo'][i][k] for i in range(len(dic_results['amarillo']))]
                flag = [a+b for a, b in zip(flag, l1)]
                final_amarillo[k] = flag    
            df_amarillo =pd.DataFrame({'round':range(rounds),'=,0':final_amarillo[0],'=,1':final_amarillo[1],'!=,0':final_amarillo[2],'!=,1':final_amarillo[3]})        

            df_total = df_azul + df_amarillo
            df_total['round'] = df_total['round'].apply(lambda x:x//2)
            df_return = [df_azul,df_amarillo,df_total]
            names = ['azul','amarillo','total']
            for i,df in enumerate(df_return):
                sns.set_color_codes("pastel")
                f, ax = plt.subplots(figsize=(12, 8))
                sns.barplot(x='round',y='!=,1',data = df,color = 'b',label = '!=,1').set_title(names[i]+',group '+str(group))
                sns.barplot(x='round',y='!=,0',data = df,color = 'r',label = '!=,0')
                sns.barplot(x='round',y='=,1',data = df,color = 'c',label = '=,1')
                sns.barplot(x='round',y='=,0',data = df,color = 'g',label = '=,0')
                ax.legend(ncol=4, loc="best", frameon=True)
                ax.set_xlabel('round')
                ax.set_ylabel('options')
                plt.savefig(os.path.join(start_point,join_name)+'/bar_action_'+str(names[i])+'_group_'+str(group)+'_'+str(study_sessions[ij])+'_'+str(join_name)+'.eps')
                plt.close(f)
    return



def bar_success(dataframes_saved):
    for ij,players_info in enumerate(dataframes_saved):
        for group in range(1,int(dataframes_saved[ij][-1][0]['id_in_subsession'])+1):
            results_azul = [[0 for ij in range(4)] for j in range(len(dataframes_saved[0][1][1]))]
            results_amarillo = [[0 for ij in range(4)] for j in range(len(dataframes_saved[0][1][1]))]
            players_info_group = [players_info[i] for i in range(len(players_info)) if players_info[i][0]['id_in_subsession'] == group]
            for player in players_info_group:
                for i in range(len(player[1])):
                    if player[0]['Marker'] == player[1]['marker_partner'].iloc[i]:
                        k = 0
                    else:
                        k = 2      
                    if player[0]['Marker'] == 'azul':

                        if i == 0:
                            results_azul[i][k+abs(1-int(player[1]['success'].iloc[i]))] += 1
                        else:
                            results_azul[i][k+abs(1-int(player[1]['success'].iloc[i])+int(player[1]['success'].iloc[i-1]))] += 1
                    elif player[0]['Marker'] == 'amarillo':
                        if i == 0:
                            results_amarillo[i][k+abs(1-int(player[1]['success'].iloc[i]))] += 1
                        else:
                            results_amarillo[i][k+abs(1-int(player[1]['success'].iloc[i])+int(player[1]['success'].iloc[i-1]))] += 1
                dic_results = {'azul':results_azul,'amarillo':results_amarillo}

            rounds = len(dic_results['azul']) ## Aquí escribín azul mais pode ser calquera
            final_azul = [0]*4
            flag = [0]*(len(dic_results['azul']))
            for k in range(4):
                l1 = [dic_results['azul'][i][k] for i in range(len(dic_results['azul']))]
                flag = [a+b for a, b in zip(flag, l1)]
                final_azul[k] = flag
            df_azul =pd.DataFrame({'round':range(rounds),'=,coord':final_azul[0],'=,uncoord':final_azul[1],'!=,coord':final_azul[2],'!=,uncoord':final_azul[3]})

            final_amarillo = [0]*4
            flag = [0]*(len(dic_results['amarillo']))
            for k in range(4):
                l1 = [dic_results['amarillo'][i][k] for i in range(len(dic_results['amarillo']))]
                flag = [a+b for a, b in zip(flag, l1)]
                final_amarillo[k] = flag    
            df_amarillo =pd.DataFrame({'round':range(rounds),'=,coord':final_amarillo[0],'=,uncoord':final_amarillo[1],'!=,coord':final_amarillo[2],'!=,uncoord':final_amarillo[3]})        

            df_total = df_azul + df_amarillo
            df_total['round'] = df_total['round'].apply(lambda x:x//2)
            df_return = [df_azul,df_amarillo,df_total]
            names = ['azul','amarillo','total']
            for i,df in enumerate(df_return):
                fig, ax = plt.subplots(figsize=(12, 8))
                sns.set_color_codes("pastel")
                sns.barplot(x='round',y='!=,uncoord',data = df,color = 'salmon',label = '!=,uncoord').set_title(names[i]+',group '+str(group))
                sns.barplot(x='round',y='!=,coord',data = df,color = 'red',label = '!=,coord')
                sns.barplot(x='round',y='=,uncoord',data = df,color = 'palegreen',label = '=,uncoord')
                sns.barplot(x='round',y='=,coord',data = df,color = 'green',label = '=,coord')
                ax.legend(ncol=4, loc="best", frameon=True)
                ax.set_xlabel('round')
                ax.set_ylabel('options')
                plt.savefig(os.path.join(start_point,join_name)+'/bar_success_'+str(names[i])+'_group_'+str(group)+'_'+str(study_sessions[ij])+'_'+str(join_name)+'.eps')
                plt.close(fig)
    
    return


def plot_semaforo(dataframes_saved,variable):
    for ij,players_info in enumerate(dataframes_saved):
        lengths = []
        for group in range(1,int(dataframes_saved[ij][-1][0]['id_in_subsession'])+1):
            a = [players_info[ik] for ik in range(len(players_info)) if int(players_info[ik][0]['id_in_subsession']) == group]
            lengths.append(len(a))
        x_ticks_length = []
        for ik2,group_length in enumerate(lengths): 
            x_ticks_length += [ik2+1]*group_length

            A = [list(players_info[k][1][variable]) for k in range(len(players_info))]
            ##Definir unha key apropiada en función do que se procure 
            #A.sort(key=lambda x:x.count(1),reverse = True)
            A = np.array(A)
            if variable == 'pay_1_val' or variable == 'pay_2_val':
                colormap = sns.color_palette("hls", 7)
            else:
                colormap = sns.color_palette('tab10',4)
            fig_n = plt.figure(i,figsize = (12,8))
            sns.heatmap(A,cbar = variable!='acción',xticklabels=False, yticklabels=x_ticks_length,cmap = colormap)
            plt.title(study_sessions[i]+','+str(variable))
            plt.xlabel('Rounds')
            plt.ylabel('Participants ID')
            plt.savefig(os.path.join(start_point,join_name)+'/'+str(variable)+'_evolution_'+str(study_sessions[ij])+'_'+str(join_name)+'.eps')
            plt.close(fig_n)

    return

def bias_success_plot(dataframes_saved):
    ##Estudo sobre a selección perante a social norm do norm_compliance 
    i = 0
    bias_pay = [0]*sum([len(dataframes_saved[i]) for i in range(len(dataframes_saved))])
    num_yellow = [0]*sum([len(dataframes_saved[i]) for i in range(len(dataframes_saved))])
    success_final = [0]*sum([len(dataframes_saved[i]) for i in range(len(dataframes_saved))])
    max_change = [0]*sum([len(dataframes_saved[i]) for i in range(len(dataframes_saved))])
    for ij,players_info in enumerate(dataframes_saved): 
        for player in players_info: 
            num_yellow[i] = list(player[2]['selectionYellow']).count(1)
            success_final[i] = player[1]['success'].iloc[-1]
            bias_pay[i] = player[0]['Bias_payoff']
            max_change[i] = max(player[1]['change_acción'])
            i +=1

    plt.figure(8026,figsize=(12,8))
    sns.relplot(x= num_yellow,y = success_final)
    plt.title('')
    plt.xlabel('Number of yellow boxes chosen')
    plt.ylabel('Success in first part')
    plt.savefig(os.path.join(start_point,join_name)+'/bias_success_'+str(study_sessions[ij])+'_'+str(join_name)+'.eps')
    plt.close()
    return

def histo_size(dataframes_saved):
    
    ##Histograms of decision
    for ij,players_info in enumerate(dataframes_saved):
        p_same = []
        p_diff = []
        group = []
        length = []
        for player in players_info:
            marker_same = player[1][player[1]['marker_partner']==player[0]['Marker']]
            marker_diff = player[1][player[1]['marker_partner']!=player[0]['Marker']]
            marker_same_1 = marker_same[marker_same['acción']==1]
            marker_diff_1 = marker_diff[marker_diff['acción']==1]
            p_same.append(len(marker_same_1)/len(marker_same))
            p_diff.append(len(marker_diff_1)/len(marker_diff))
            group.append(player[0].id_in_subsession)
            length.append([players[0].id_in_subsession == player[0].id_in_subsession for players in players_info].count(True))
        plt.figure(8001,figsize = (12,8))
        sns.scatterplot(x=p_diff, y=p_same,hue=length)
        sns.rugplot(x=p_diff, y=p_same)
        plt.title('Histogram-color indicates size of group')
        plt.xlabel(r'$p(=,\Delta)$')
        plt.ylabel(r'$p(\neq,\Delta)$')
        plt.savefig(os.path.join(start_point,join_name)+'/histo_size_'+str(study_sessions[ij])+'_'+str(join_name)+'.eps')
        plt.close()
    
    
    ##Statistics about correlation between variables
    pearson_r = scipy.stats.pearsonr(p_diff, p_same) 
    spearman_rho =  scipy.stats.spearmanr(p_diff,p_same)   
    kendall_tau = scipy.stats.kendalltau(p_diff,p_same)  
    print('Pearson R is {:3f}, with a p-value of {:3f}'.format(pearson_r[0],pearson_r[1]))
    print('Spearman rho is {:3f}, with a p-value of {:3f}'.format(spearman_rho[0],spearman_rho[1]))
    print('Kendall tau is {:3f}, with a p-value of {:3f}'.format(kendall_tau[0],kendall_tau[1])) 
    
    return


def histo_group(dataframes_saved):
    
    ##Histograms of decision
    for ij,players_info in enumerate(dataframes_saved):
        p_same = []
        p_diff = []
        group = []
        length = []
        for player in players_info:
            marker_same = player[1][player[1]['marker_partner']==player[0]['Marker']]
            marker_diff = player[1][player[1]['marker_partner']!=player[0]['Marker']]
            marker_same_1 = marker_same[marker_same['acción']==1]
            marker_diff_1 = marker_diff[marker_diff['acción']==1]
            p_same.append(len(marker_same_1)/len(marker_same))
            p_diff.append(len(marker_diff_1)/len(marker_diff))
            group.append(player[0].id_in_subsession)
            length.append([players[0].id_in_subsession == player[0].id_in_subsession for players in players_info].count(True))
        plt.figure(8001,figsize = (12,8))
        sns.scatterplot(x=p_diff, y=p_same,hue=group)
        sns.rugplot(x=p_diff, y=p_same)
        plt.title('Histogram-color indicates group')
        plt.xlabel(r'$p(=,\Delta)$')
        plt.ylabel(r'$p(\neq,\Delta)$')
        plt.savefig(os.path.join(start_point,join_name)+'/histo_group_'+str(study_sessions[ij])+'_'+str(join_name)+'.eps')
        plt.close()
    
    
    ##Statistics about correlation between variables
    pearson_r = scipy.stats.pearsonr(p_diff, p_same) 
    spearman_rho =  scipy.stats.spearmanr(p_diff,p_same)   
    kendall_tau = scipy.stats.kendalltau(p_diff,p_same)  
    print('Pearson R is {:3f}, with a p-value of {:3f}'.format(pearson_r[0],pearson_r[1]))
    print('Spearman rho is {:3f}, with a p-value of {:3f}'.format(spearman_rho[0],spearman_rho[1]))
    print('Kendall tau is {:3f}, with a p-value of {:3f}'.format(kendall_tau[0],kendall_tau[1])) 
    
    return

def histo_previous_success(dataframes_saved):
    
    ##Histograms of decision
    for ij,players_info in enumerate(dataframes_saved):
        p_same = []
        p_diff = []
        group = []
        length = []
        for player in players_info:
            acción_1 = player[1][player[1]['acción']==1]
            acción_2 = player[1][player[1]['acción']!=1]
            acción_1_ref = [acción_1['payoff'].iloc[i] for i in range(1,len(acción_1)) if (acción_1['payoff'].iloc[i-1] == 1.5)]
            acción_2_ref = [acción_2['payoff'].iloc[i] for i in range(1,len(acción_2)) if (acción_2['payoff'].iloc[i-1] == 1.5)]
            if len(acción_1) > 0:
                p_same.append(len(acción_1_ref)/len(acción_1))
            else:
                p_same.append(0)
            if len(acción_2) > 0:
                p_diff.append(len(acción_2_ref)/len(acción_2))
            else:
                p_diff.append(0)
            group.append(player[0].id_in_subsession)
            length.append([players[0].id_in_subsession == player[0].id_in_subsession for players in players_info].count(True))
        plt.figure(8001,figsize = (12,8))
        sns.scatterplot(x=p_diff, y=p_same,hue=group)
        sns.rugplot(x=p_diff, y=p_same)
        plt.title('Reinforcement of an action given a previous success')
        plt.xlabel(r'$p(ac_2,ref)$')
        plt.ylabel(r'$p(ac_1,ref)$')
        plt.savefig(os.path.join(start_point,join_name)+'/histo_previous_success_'+str(study_sessions[ij])+'_'+str(join_name)+'.eps')
        plt.close()
    
    ##Statistics about correlation between variables
    pearson_r = scipy.stats.pearsonr(p_diff, p_same) 
    spearman_rho =  scipy.stats.spearmanr(p_diff,p_same)   
    kendall_tau = scipy.stats.kendalltau(p_diff,p_same)  
    print('Pearson R is {:3f}, with a p-value of {:3f}'.format(pearson_r[0],pearson_r[1]))
    print('Spearman rho is {:3f}, with a p-value of {:3f}'.format(spearman_rho[0],spearman_rho[1]))
    print('Kendall tau is {:3f}, with a p-value of {:3f}'.format(kendall_tau[0],kendall_tau[1]))  
    
    return
                
                
#Histo with groups but only with a certain number of last values, to show the final behavior.
def histo_group_end(dataframes_saved,valor):
    for ij,players_info in enumerate(dataframes_saved):
        ##Histograms of decision
        p_same = []
        p_diff = []
        group = []
        length = []
        for player in players_info:
            marker_same = player[1][player[1]['marker_partner']==player[0]['Marker']]
            marker_diff = player[1][player[1]['marker_partner']!=player[0]['Marker']]
            marker_same = marker_same[marker_same["round_number"]>valor]
            marker_diff = marker_diff[marker_diff["round_number"]>valor]
            marker_same_1 = marker_same[marker_same['acción']==1]
            marker_diff_1 = marker_diff[marker_diff['acción']==1]
            p_same.append(len(marker_same_1)/len(marker_same))
            p_diff.append(len(marker_diff_1)/len(marker_diff))
            group.append(player[0].id_in_subsession)
            length.append([players[0].id_in_subsession == player[0].id_in_subsession for players in players_info].count(True))
        plt.figure(8001,figsize = (12,8))
        sns.scatterplot(x=p_diff, y=p_same,hue=group)
        sns.rugplot(x=p_diff, y=p_same)
        plt.title('Histogram-color indicates group')
        plt.xlabel(r'$p(=,\Delta)$')
        plt.ylabel(r'$p(\neq,\Delta)$')
        plt.savefig(os.path.join(start_point,join_name)+'/histo_group_end_'+str(study_sessions[ij])+'_'+str(join_name)+'.eps')
        plt.close()
    
    return

def moving_coordination_rate(dataframes_saved,window):
    for ij,players_info in enumerate(dataframes_saved):
        rounds_1 = int(max(df_1['subsession.round_number'].unique()))
        n_grupos = int(dataframes_saved[-1][-1][0]['id_in_subsession']+1)
        success_rate = [[0 for i in range(rounds_1)] for i in range(n_grupos)]
        for group in range(1,n_grupos):
            players_info_group = [players_info[i] for i in range(len(players_info)) if players_info[i][0]['id_in_subsession'] == group]
            for ronda in range(rounds_1):
                     success_rate[group-1][ronda] = len([players_info_group[j][1]['payoff'].iloc[ronda] for j in range(len(players_info_group))
                          if players_info_group[j][1]['payoff'].iloc[ronda]==1])/len(players_info_group)
        for i in range(n_grupos-1):
            plt.figure(8002,figsize = (12,8))
            plt.title("Moving average of the coordination rate, window ="+str(window))
            plt.plot(moving_average(success_rate[i],10),label = "Grupo:" +str(i))
            plt.xlabel('Round')
            plt.ylabel('Coordination rate')
            plt.legend()
            plt.savefig(os.path.join(start_point,join_name)+'/moving_coordination_rate_'+str(study_sessions[ij])+'_'+str(join_name)+'.eps')       
        plt.close()
    return

def coordination_distribution(dataframes_saved,valor):
    for ij,players_info  in enumerate(dataframes_saved):
        rounds_1 = int(max(df_1['subsession.round_number'].unique()))
        n_grupos = int(dataframes_saved[-1][-1][0]['id_in_subsession']+1)
        success_rate = [[] for i in range(n_grupos-1)]
        for group in range(1,n_grupos):
            players_info_group = [players_info[i] for i in range(len(players_info)) if players_info[i][0]['id_in_subsession'] == group]
            for ij2,player in enumerate(players_info_group): 
                success_rate[group-1].append((player[1]['success'].iloc[-1] - player[1]["success"].iloc[valor])/(player[1]['round_number'].iloc[-1] - player[1]["round_number"].iloc[valor]))
        fig_1, ax = plt.subplots(n_grupos-1,1, figsize=(8,16))
        for i in range(len(success_rate)):
            if i == 0: 
                ax[i].title.set_text("Coordination rate distribution, ronda fin transitorio ="+str(valor))
            ax[i].hist(success_rate[i],label = "Grupo:" +str(i),alpha=0.5,bins=[i/10 for i in range(3,11)])
            ax[i].set_ylabel("Grupo:"+str(i+1))
        plt.savefig(os.path.join(start_point,join_name)+'/coordination_distribution_'+str(study_sessions[ij])+'_'+str(join_name)+'.eps') 
        plt.close()
    return

def change_action_distribution(dataframes_saved,valor):
    for ij,players_info  in enumerate(dataframes_saved):
        rounds_1 = int(max(df_1['subsession.round_number'].unique()))
        n_grupos = int(dataframes_saved[-1][-1][0]['id_in_subsession']+1)
        lista = [[] for i in range(n_grupos-1)]
        for group in range(1,n_grupos):
            players_info_group = [players_info[i] for i in range(len(players_info)) if players_info[i][0]['id_in_subsession'] == group]
            for ij2,player in enumerate(players_info_group): 
                player_transi = player[1][player[1]["round_number"]<valor]
                a = player_transi[(player_transi["change_marker"] == 1)&(player_transi["change_acción"] == 1)]
                b = player_transi[(player_transi["change_marker"] == 1)]
                lista[group-1].append(len(a)/len(b))
        fig_2, ax = plt.subplots(n_grupos-1,1, figsize=(8,16))
        for i in range(len(lista)):
            if i == 0: 
                ax[i].title.set_text("Change action distribution, ronda fin transitorio ="+str(valor))
            ax[i].hist(lista[i],label = "Grupo:" +str(i),alpha=0.5,bins=[i/10 for i in range(0,11)])
            ax[i].set_ylabel("Grupo:"+str(i+1))
        plt.savefig(os.path.join(start_point,join_name)+'/change_action_distribution_'+str(study_sessions[ij])+'_'+str(join_name)+'.eps') 
        plt.close()
    
    return

def aspiration_correlation(dataframes_saved):
    for ij,players_info in enumerate(dataframes_saved):
        pay_1_success= []
        pay_2_success = []
        av_success = []
        period = []
        for player in players_info:
            j2 = 0
            for j in range(14,74,15):
                pay_1_success.append(player[1]["pay_1_val"].iloc[j]-player[1]["pay_1_val"].iloc[j2])
                pay_2_success.append(player[1]["pay_2_val"].iloc[j]-player[1]["pay_1_val"].iloc[j2])
                av_success.append(np.average(player[1][:j]["success"])-np.average(player[1][:j2]["success"]))
                period.append(j)
                j2 = j

        plt.figure(8023,figsize = (12,8))
        sns.scatterplot(x=av_success, y=pay_1_success,hue=period)
        plt.title('Change in aspirations related to average success')
        plt.xlabel(r'Average success')
        plt.ylabel(r'Low aspiration change')
        plt.legend(title="Initial round for average")
        plt.savefig(os.path.join(start_point,join_name)+'/pay_1_success_'+str(study_sessions[ij])+'_'+str(join_name)+'.eps')
        plt.close()

        plt.figure(8024,figsize = (12,8))
        sns.scatterplot(x=av_success, y=pay_2_success,hue=period)
        plt.title('Change in aspirations related to average success')
        plt.xlabel(r'Average success')
        plt.ylabel(r'High aspiration change')
        plt.legend(title="Initial round for average")
        plt.savefig(os.path.join(start_point,join_name)+'/pay_2_success_'+str(study_sessions[ij])+'_'+str(join_name)+'.eps')
        plt.close()
    return
```

```python
#Métodos de listas
import numpy as np

def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w
```

```python
#Métodos de inclusión de variables
from scipy import stats

def include_new_groups(dataframes_saved,study_sessions):
    if ("7/6/2021" in study_sessions) and ("24/6/2021" in study_sessions):
        extra_session_index = "24/6/2021".index(study_sessions)
        init_session_index = "7/6/2021".index(study_sessions)
        init_session = dataframes_saved[init_session_index] 
        new_group = [init_session[i] for i in range(len(init_session)) if init_session[i][0]['id_in_subsession'] == 1]
        for player in new_group:
            player[0]["id_in_subsession"] = 6
        dataframes_saved[extra_session_index].append(new_group)
        del study_sessions[init_session_index]
    return dataframes_saved,study_sessions
        

def include_aspiration(dataframes_saved):
    for players_info in dataframes_saved:
        for player in players_info:
            player_vals = [list(abs(player[1].pay_1_val-4)),list(abs(player[1].pay_2_val-4))]
            player_vals_2 = [list(player[1].pay_1_val-4),list(player[1].pay_2_val-4)]
            player_vals = list(zip(player_vals[0],player_vals[1]))
            player_vals_2 = list(zip(player_vals_2[0],player_vals_2[1]))
            stim = [0 for i in range(2)]
            aspiration = [0]*len(player[1])
            for i in range(len(player_vals)):
                if player_vals_2[i][0]<player_vals_2[i][1]:
                    max_index = player_vals[i].index(max(player_vals[i]))
                    stim[max_index] = player_vals_2[i][max_index]/abs(player_vals[i][max_index])
                    stim[abs(1-max_index)] = player_vals_2[i][abs(1-max_index)]/abs(player_vals[i][max_index])
                    aspiration[i] = (stim[1]*1 - stim[0]*1.5)/(stim[1]-stim[0])
                else: 
                    aspiration[i] = 10

            player[1]['Aspiration'] = aspiration

def include_semaforo(dataframes_saved): 
    ###Include aspirations and include +/-acción, pra ver como evolucionan as estratexias individuais
    for players_info in dataframes_saved:
        for player in players_info: 
            semaforo = [0]*len(player[1])
            change_accion = [0]*len(player[1])
            change_marker = [0]*len(player[1])
            for i in range(1,len(player[1])):
                if player[1]['acción'].iloc[i] != player[1]['acción'].iloc[i-1]:
                    change_accion [i] =  1
                if player[1]["marker_partner"].iloc[i] != player[1]["marker_partner"].iloc[i-1]:
                    change_marker[i] = 1
                if i == 0: 
                    semaforo[i] = player[1]['acción'].iloc[i]*(-1)**(1+player[1]['success'].iloc[i])
                else: 
                    semaforo[i] = player[1]['acción'].iloc[i]*(-1)**(1+player[1]['success'].iloc[i]-player[1]['success'].iloc[i-1])
            player[1]['semaforo'] = semaforo
            player[1]['change_acción'] = change_accion
            player[1]['change_marker'] = change_marker
    return 


def normality_trust(dataframes_saved):
    dic_bias = {'amarillo':532,'azul':594}
    total_len = sum([len(dataframes_saved[i]) for i in range(len(dataframes_saved))])
    distancias_igual = [0 for ij in range(total_len)]
    distancias_diff =  [0 for ij in range(total_len)]
    k = 0
    for players_info in dataframes_saved:
        for i in range(len(players_info)):
            for mode in ['igual','diferente']:
                if mode == 'igual':
                    distancias_igual[k] = abs(dic_bias[players_info[i][0]['Marker']] - players_info[i][0]['Bias'])
                else: 
                    if players_info[i][0]['Marker'] == 'amarillo': maskarita = 'azul'
                    if players_info[i][0]['Marker'] == 'azul': maskarita = 'amarillo'
                    distancias_diff[k] = abs(dic_bias[maskarita] - players_info[i][0]['Bias'])
            k+=1


    nearer = [(distancias_diff[i]-distancias_igual[i])/(distancias_igual[i]+distancias_diff[i]) for i in range(len(distancias_igual))]
    bias_plots = pd.DataFrame({'distance to same marker reference':distancias_igual,
                              'distance to different marker reference':distancias_diff,
                              'trust': nearer
                              })
    ks_test = stats.ks_2samp(np.array(bias_plots['distance to same marker reference']),np.array(bias_plots['distance to different marker reference']))
    shap_test = stats.shapiro(np.array(bias_plots['trust']))
    print('The Kolmogorov statistic is {:.3f} and its p-value is {:.3f} and the Shapiro test statistic is {:.3f} and its p-value is {:.3f}'.format(ks_test[0],ks_test[1],shap_test[0],shap_test[1]))
    return
```
