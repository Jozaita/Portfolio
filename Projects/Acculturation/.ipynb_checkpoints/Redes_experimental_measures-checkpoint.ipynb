```python

from numpy import nan
import pandas as pd 
import networkx as nx
from methods import *
import numpy as np
import matplotlib 
from matplotlib import pyplot as plt
import ast
#Importamos o arquivo de contactos 
contactos=pd.read_csv(r'/home/juan/Python/Acculturation/Contactos.csv',low_memory=False)
df=pd.read_csv(r'/home/juan/Python/Acculturation/all_data_clean.csv',low_memory=False)

del contactos['Unnamed: 0']
del df['Unnamed: 0']
```

```python
#contactos=contactos.dropna()
contactos['Alter']=contactos['Alter'].astype(int)
contactos['Alter2']=contactos['Alter2'].astype(int)
#Atopa o número de suxeitos
k=1
for i in range(1,len(contactos)):
    if (contactos['sub/num'][i]!=contactos['sub/num'][i-1]):
        k=k+1
sujetos=k
print(k)
```

```python
#Creación de gráficas e variáveis asociadas
graficas=[0]*sujetos
avdeg=[0]*sujetos
betw=[0]*sujetos
betw_2=[0]*sujetos
closs=[0]*sujetos
assort=[0]*sujetos
clustering=[0]*sujetos
clustering_2=[0]*sujetos
load=[0]*sujetos
load_2=[0]*sujetos
size=[0]*sujetos
comp=[0]*sujetos
ori=[0]*sujetos
num=[0]*sujetos
res=[0]*sujetos
closnac=[0]*sujetos
closnonac=[0]*sujetos
numnac=[0]*sujetos
mu=[0]*sujetos
numnonac=[0]*sujetos
vect=[[0]*5 for i in range(sujetos)]
origins=contactos['sub/origin'].unique()
k=0
for i in origins:
    nummax=max(contactos[contactos['sub/origin']==i]['sub/num'])
    for j in range(1,nummax+1):
        graficas[k]=nx.Graph()
        ###Add graph attributes from df
        data=contactos[((contactos['sub/num']==j)&(contactos['sub/origin']==i))]
        if (len(data)>0):
            ##Mus
            datamu=df['Clos'][(df['ego/origin']==i)&(df['ego/num']==j)].value_counts()
            for m in range(0,len(datamu)):
                vect[k][datamu.index[m]-1]= datamu[datamu.index[m]]
            ##Parametros de rede
            edges=list(zip(data['Alter'],data['Alter2'],data['Value']))
            graficas[k].add_weighted_edges_from(edges,'Value')
            degrees=[val for (node, val) in graficas[k].degree()]
            #Versión sin ponderar
            betw_2[k]=sum(nx.betweenness_centrality(graficas[k]).values())/len(nx.betweenness_centrality(graficas[k]).values())
            #Versión ponderada
            betw[k]=sum(nx.betweenness_centrality(graficas[k],weight='Value').values())/len(nx.betweenness_centrality(graficas[k]).values())
            closs[k]=sum(nx.closeness_centrality(graficas[k]).values())/len(nx.closeness_centrality(graficas[k]).values())
            #Versión ponderada
            load[k]=sum(nx.load_centrality(graficas[k],weight='Value').values())/len(nx.load_centrality(graficas[k]).values())
            #Versión sin ponderar
            load_2[k]=sum(nx.load_centrality(graficas[k]).values())/len(nx.load_centrality(graficas[k]).values())
            assort[k]=nx.degree_assortativity_coefficient(graficas[k])
            size[k]= len(max(nx.connected_components(graficas[k]), key=len))/len(graficas[k].nodes())
            #Versión ponderada
            clustering[k]=nx.average_clustering(graficas[k],weight='Value')
            #Versión sin ponderar
            clustering_2[k]=nx.average_clustering(graficas[k])
            avdeg[k]=sum(degrees)/len(degrees)
            comp[k]=nx.number_connected_components(graficas[k])
            closnac[k]=df['Clos'][((df['ego/residence']==df['alter/origin'])
                                   &(df['ego/num']==j)&(df['ego/origin']==i))].mean()
            numnac[k]=df['Clos'][((df['ego/residence']==df['alter/origin'])
                                   &(df['ego/num']==j)&(df['ego/origin']==i))].count()
            closnonac[k]=df['Clos'][((df['ego/origin']==df['alter/origin'])
                                   &(df['ego/num']==j)&(df['ego/origin']==i))].mean()
            numnonac[k]=df['Clos'][((df['ego/origin']==df['alter/origin'])
                                   &(df['ego/num']==j)&(df['ego/origin']==i))].count()
            ori[k]=i
            num[k]=j
            res[k]=data['sub/residence'][((data['sub/origin']==i)&(data['sub/num']==j))].unique()[0]
            k=k+1
        else: print(i,j)


#for i in range(len(graficas)):
#    graficas[i]=nx.Graph()
#    graph=contactos[contactos['sub/num']==sujetos[i]]
#    edges=list(zip(graph['Alter'],graph['Alter2'],graph['Value']))    
#    graficas[i].add_weighted_edges_from(edges)
#    degrees=[val for (node, val) in graficas[i].degree()]
#    betw[i]=sum(nx.betweenness_centrality(graficas[i]).values())/len(nx.betweenness_centrality(graficas[i]).values())
#    avdeg[i]=sum(degrees)/len(degrees)
#    comp[i]=nx.number_connected_components(graficas[i])gr
```

```python
#función pra determinar o réxime
def rex(x,y):
    if ((x < 0) & (y < 0)): z='Inverted'
    elif ((x > 0) & (y > 0)): z='Standard'
    else : z='Unclear'
    return z
```

```python
#Creación de dataframe
redes=pd.DataFrame(data={'Subject origin':ori,'Subject residence':res,'Subject ID':num,'Vect':vect,'Graphs':graficas,
                         'Average degree':avdeg,'Betweenness':betw,'Closeness':closs,'Load centrality':load,
                         'Assortativity':assort,'Clustering':clustering,
                         'Number components':comp,'Size largest component':size,
                         'Closeness residence':closnac,'Closeness origin':closnonac,'Number residence':numnac,
                         'Number origin':numnonac})
redes['Fitted']=redes['Vect'].apply(lambda x:Individual(x).fit_model())
redes[['Mu','Mu-','Mu+']] = pd.DataFrame(redes.Fitted.values.tolist(), index= redes.index)


len(redes)
```

```python
regime=[0]*sujetos
for i in range(len(redes)):
    regime[i]=rex(redes['Mu-'][i],redes['Mu+'][i])
redes['Regime']=regime
del redes['Fitted']
del redes['Mu-']
del redes['Mu+']
```

```python
redes.fillna(0,inplace=True)

redes=redes[['Subject ID','Subject origin','Subject residence','Mu','Regime','Average degree','Betweenness',
             'Closeness','Load centrality','Assortativity','Clustering',
             'Number components','Size largest component',
             'Closeness residence','Number residence','Closeness origin','Number origin','Graphs']]
len(redes)
```

```python

del df['alter/origin']
del df['alter/residence']
del df['alter/num']
del redes['Graphs']
df.rename(columns={'ego/num':'Subject ID','ego/origin':'Subject origin','ego/residence':'Subject residence'},inplace=True)
df =df.drop_duplicates(['Subject origin','Subject ID'])
redes2= pd.merge(redes, df,how='left',on=['Subject ID','Subject origin','Subject residence'])


import datetime
FMIG2=[0]*len(redes2)
#redes2['FMIG2']=datetime.datetime.now().year-redes2['FMIG']
for i in range(len(redes2)):
    if (redes2['FMIG'].iloc[i]<0.5): 
        FMIG2[i]=0
    elif (redes2['FMIG'].iloc[i]<1500) and (redes2['FMIG'].iloc[i]>0.5):
        FMIG2[i]=datetime.datetime.now().year-redes2['FMIG'].iloc[i]

redes2['FMIG2']=FMIG2

redes2.rename(columns={'Subject ID':'Subject_num','Subject origin':'Subject_origin','Subject residence':'Subject_residence'},inplace=True)
redes2.columns = redes2.columns.str.replace(' ', '_')
redes2.to_csv('Redes_2.csv')
```
