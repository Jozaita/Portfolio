```python
from sklearn.datasets import make_circles

n_samples = 1000

#Crear círculos
X,Y = make_circles(n_samples,
                  noise=0.03,
                  random_state=42)
```

```python
#Modelaje

import tensorflow as tf

tf.random.set_seed(42)

model_1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

model_1.compile(loss=tf.keras.losses.BinaryCrossentropy(),
               optimizer=tf.keras.optimizers.Adam(),
               metrics = ["accuracy"]
               )
model_1.fit(X,Y,epochs=5)
```

```python
model_1.fit(X,Y,epochs=200,verbose=0)
model_1.evaluate(X,Y)
```

```python
#Baja precisión, añadir otra capa
tf.random.set_seed(42)

model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(1),
    tf.keras.layers.Dense(1)
])

model_2.compile(loss=tf.keras.losses.BinaryCrossentropy(),
               optimizer=tf.keras.optimizers.SGD(),
               metrics=["accuracy"])

model_2.fit(X,Y,epochs=100,verbose=0)
```

```python
model_2.evaluate(X,Y)
```

```python
#La precisión no mejora

tf.random.set_seed(42)

model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model_3.compile(loss=tf.keras.losses.BinaryCrossentropy(),
               optimizer=tf.keras.optimizers.Adam(),
                metrics=["accuracy"])
                
                
model_3.fit(X,Y,epochs=100,verbose=0)
```

```python
model_3.evaluate(X,Y)
```

```python
#Visualizar predicciones

import numpy as np

def plot_decision_boundary(model,X,Y):
    
    #Definir los ejes del plot y crear un meshgrid
    x_min,x_max =X[:,0].min()-0.1,X[:,0].max()+0.1
    y_min,y_max = X[:,1].min()-0.1,X[:,1].max()+0.1
    xx,yy = np.meshgrid(np.linspace(x_min,x_max,100),
    np.meshgrid(np.linspace(y_min,y_max,100)))
    #Crear predicciones
    x_in = np.c_[xx.ravel(),yy.ravel()]
    
    y_pred = model.predict(x_in)
    
    #Muilticlass
    if len(y_pred[0])>1:
        print("Multiclass classification")
        #Reshapear las predicciones
        y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
    else:
        print("Binary classification")
        y_pred = np.round(y_pred).reshape(xx.shape)
        
    plt.contourf(xx,yy,y_pred,cmap=plt.cm.RdYlBu,alpha=0.7)
    plt.scatter(X[:,0],X[:,1],c=Y,s=40,cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
```

```python
plot_decision_boundary(model=model_3,X=X,Y=Y)
```

```python
#Comprobar si este modelo podría ser usado para un problema de regresión

tf.random.set_seed(42)

X_regression = tf.range(0,1000,5)
Y_regression = tf.range(100,1100,5)
#SPlit entre training y test

X_reg_train = X_regression[:150]
X_reg_test = X_regression[150:]
Y_reg_train = Y_regression[:150]
Y_reg_test = Y_regression[150:]

model_3.fit(X_reg_train,Y_reg_train,epochs=100)
```

```python
#Recompilar para una regresión 
tf.random.set_seed(42)

model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model_3.compile(loss=tf.keras.losses.mae,
               optimizer=tf.keras.optimizers.Adam(),
               metrics=["mae"])

model_3.fit(X_reg_train,Y_reg_train,epochs=100)
```

```python
#Hacer predicciones y ploteat
Y_reg_preds = model_3.predict(X_reg_test)
plt.figure(figsize=(10,7))
plt.scatter(X_reg_train,Y_reg_train,c ="b",label="training data")
plt.scatter(X_reg_test,Y_reg_test,c="g",label = "test data")
plt.scatter(X_reg_test,Y_reg_preds,c="r",label="predictions")
plt.legend()
```

```python
#Evaluando y mejorando el modelo de clasificación
#Crear un training y un test set

X_train,Y_train = X[:800],Y[:800]
X_test,Y_test = X[800:],Y[800:]

X_train.shape,Y_train.shape,X_test.shape,Y_test.shape
```

```python
tf.random.set_seed(42)

model_8 = tf.keras.Sequential([
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(1,activation="sigmoid")
])

model_8.compile(loss=tf.keras.losses.BinaryCrossentropy(),
               optimizer=tf.keras.optimizers.Adam(learning_rate=10e-3),
               metrics=["accuracy"])

history = model_8.fit(X_train,Y_train,epochs=20)
```

```python
#Modificar el learning rate para encontrar el óptimo
#Pasos:
#Crear un callback, esto es, una pieza extra de funcionalidad
#que puedes añadir mientras entrena.
#Otro modelo
#Una curva de error modificada. 
```

```python
#Crear el modelo
tf.random.set_seed(42)

model_9 = tf.keras.Sequential([
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(4,activation="relu"),
    tf.keras.layers.Dense(1,activation="sigmoid")
])

model_9.compile(loss=tf.keras.losses.BinaryCrossentropy(),
               optimizer=tf.keras.optimizers.Adam(),
               metrics=["accuracy"])

#Crear el callback
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epoch:1e-4*10**(epoch/20))


history_9 = model_9.fit(X_train,
                       Y_train,
                       epochs=100,
                       callbacks=[lr_scheduler])
```

```python
#Mirar la historia
pd.DataFrame(history_9.history).plot(figsize=(10,7),xlabel="epochs")
```

```python
plt.semilogx(history_9.history["lr"],history_9.history["loss"])
plt.xlabel("LR")
plt.ylabel("loss")
```
