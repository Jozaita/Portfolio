```python
#Visualizar predicciones

import numpy as np

def plot_decision_boundary(model,X,Y):
    
    #Definir los ejes del plot y crear un meshgrid
    x_min,x_max =X[:,0].min()-0.1,X[:,0].max()+0.1
    y_min,y_max = X[:,1].min()-0.1,X[:,1].max()+0.1
    xx,yy = np.meshgrid(np.linspace(x_min,x_max,100),
    np.meshgrid(np.linspace(y_min,y_max,100)))
    #Crear predicciones
    x_in = np.c_[xx.ravel(),yy.ravel()]
    
    y_pred = model.predict(x_in)
    
    #Muilticlass
    if len(y_pred[0])>1:
        print("Multiclass classification")
        #Reshapear las predicciones
        y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
    else:
        print("Binary classification")
        y_pred = np.round(y_pred).reshape(xx.shape)
        
    plt.contourf(xx,yy,y_pred,cmap=plt.cm.RdYlBu,alpha=0.7)
    plt.scatter(X[:,0],X[:,1],c=Y,s=40,cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
```

```python
plot_decision_boundary(model=model_3,X=X,Y=Y)
```

```python
#Hacer más bonita la matriz de confusión

import itertools
from sklearn.metrics import confusion_matrix
figsize = (10,10)

def make_confusion_matrix(y_true,y_pred,classes = None,figsize=(10,10),text_size=15):
#Matriz de confusión
    cm = confusion_matrix(y_true,tf.math.round(y_pred))
    cm_norm = cm.astype("float")/cm.sum(axis=1)[:,np.newaxis]
    n_classes = cm.shape[0]

    #Más bonita
    fig,ax = plt.subplots(figsize=figsize)
    cax = ax.matshow(cm,cmap=plt.cm.Blues)
    fig.colorbar(cax)

    #Crear clases


    if classes : 
        labels = classes
    else:
        labels = np.arange(cm.shape[0])

    ax.set(title="Confusion Matrix",
          xlabel="Predicted label",
          ylabel="True label",
          xticks=np.arange(n_classes),
          yticks = np.arange(n_classes),
          xticklabels=labels,
          yticklabels=labels)

    ax.xaxis.set_label_position("bottom")
    ax.xaxis.tick_bottom()

    ax.yaxis.label.set_size(text_size)
    ax.xaxis.label.set_size(text_size)
    ax.title.set_size(text_size)

    threshold = (cm.max() + cm.min())/2

    for i,j in itertools.product(range(cm.shape[0]),range(cm.shape[1])):
        plt.text(j,i,f"{cm[i,j]}({cm_norm[i,j]*100:.1f}%)",
                horizontalalignment="center",
                color="white" if cm[i,j] > threshold else "black",
                size=text_size)
```

```python
#Crear una función para plotear una imagen aleatoria, hacer 
#predicción y comparar las etiquetas
import random 

def plot_random_image(model,images,true_labels,classes):
    """
    Picks a random image, plots it and labels it with prediction"""
    i = random.randint(0,len(images))
    target_image = images[i]
    pred_probs = model.predict(target_image.reshape(1,28,28))
    pred_label = classes[pred_probs.argmax()]
    true_label = classes[true_labels[i]]
    
    plt.imshow(target_image,cmap =plt.cm.binary)
    
    #Pijada
    if pred_label == true_label:
        color = "g"
        
    else: 
        color = "r"
        
    plt.xlabel("Pred: {}{:2.0f}%(True:{})".format(pred_label,
                                                 100*tf.reduce_max(pred_probs),
                                                 true_label),
              color=color)
```

```python
def show_class(guess_class,n,class_names,train_data_norm,test_labels):
    guess_class = "Dress"
    guess_index = class_names.index(guess_class)

    t_1 = [test_data_norm[i] for i in range(len(test_data_norm)) if test_labels[i] == guess_index]


    sample = random.sample(range(0,len(t_1)),n)
    set_1 = [t_1[sample[i]] for i in range(len(sample))]
    y_probs = [model_18.predict(set_1[i].reshape(1,28,28)) for i in range(len(set_1))]
    y_preds = [tf.argmax(y_probs[i],axis=1) for i in range(len(y_probs))]

    num_images = n
    plt.figure(figsize=(3*num_images, 3*num_images))
    for i in range(num_images):
        plt.subplot(num_rows,2*num_cols,2*i+1)
        plt.imshow(t_1[sample[i]],cmap =plt.cm.binary)

        #Pijada
        if class_names[int(y_preds[i].numpy())] == guess_class:
            color = "g"

        else: 
            color = "r"

        plt.xlabel("Pred: {}{:2.0f}%".format(y_preds[i],
                                                     100*tf.reduce_max(y_probs[i])),
                  color=color)
```

```python
show_class("Shirt",6,class_names,train_data_norm,test_labels)
```
