## Seaborn 

Seaborn is a library, one of the most popular masks for matplotlib. It provides the functionality that is mainly used in scientific literature and data science. 

Apart from a wide variety of predefined plots, that can be implemented directly in matplotlib axis, but also a more visual aesthetics.

___
## Imports

```python
import seaborn as sns
```

### Datasets

Seaborn provides a variety of datasets in order to test the different plots

```python
tips = sns.load_dataset('tips')
tips
```

## distplot

The distplot shows the distribution of a univariate set of observations.

```python
sns.displot(tips['total_bill'],)
```

## jointplot

jointplot() allows you to basically match up two distplots for bivariate data. With your choice of what **kind** parameter to compare with: 
* “scatter” 
* “reg” 
* “resid” 
* “kde” 
* “hex”

```python
sns.jointplot(x='total_bill',y='size',data=tips,kind='kde')
```

### KDE plot and rug plot 

Kernel density functions are functions that represent the inference of a probability distribution. It is constructed from a sum of gaussian functions. 

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

### Categorical plots 

If variables can take just a limited amount of values. 

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='viridis')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,)
```

# Matrix Plots

Matrix plots allow you to plot data as color-encoded matrices and can also be used to indicate clusters within the data. Similar to imshow,pcolor, pcolormesh in matplotlib.

```python
tips.corr()
sns.heatmap(tips.corr(),annot=True,)
```

## Regression Plots

**lmplot** allows you to display linear models, but it also conveniently allows you to split up those plots based off of features, as well as coloring the hue based off of features.

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

### Controlling the aesthetics

```python
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips,palette='deep')
```

### Spine Removal

```python
sns.countplot(x='sex',data=tips)
sns.despine()
```

### Size and Aspect

```python
# Non Grid Plot
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips,)
```

```python
# Grid Type Plot
sns.lmplot(x='total_bill',y='tip',size=2,aspect=4,data=tips)
```

### Scale and Context

The set_context() allows you to override default parameters:

```python
sns.set_context("poster",font_scale=1)
sns.countplot(x='sex',data=tips,palette='coolwarm')
```
